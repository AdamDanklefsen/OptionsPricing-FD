# Minimal Windows demo: call C++ (pybind11) from Jupyter

This notebook will:
1. Install build-time Python packages if necessary
2. Build the C++ extension in-place using setup.py
3. Import the extension and call `hello()` and the placeholder `american_solver_stub()`

Run the notebook from the venv where you installed pybind11 (activate `\.venv\Scripts\Activate.ps1`).

```python
import sys
import subprocess
import importlib
import os

python = sys.executable
print('Python used:', python)

print('Installing build-time deps (pybind11, setuptools, wheel) if missing...')
subprocess.check_call([python, '-m', 'pip', 'install', '--upgrade', 'pip'])
subprocess.check_call([python, '-m', 'pip', 'install', 'pybind11', 'setuptools', 'wheel'])

print('Building extension (setup.py build_ext --inplace)')
subprocess.check_call([python, 'setup.py', 'build_ext', '--inplace'])
print('Build finished')
```

```python
try:
    import cpphello
except Exception as e:
    print('Failed to import cpphello:', e)
    raise

print(cpphello.hello())
print('american_solver_stub(100, 90, 0.01, 0.2, 1.0) ->', cpphello.american_solver_stub(100.0, 90.0, 0.01, 0.2, 1.0))
```
